```python
order_item_df['shipping_limit_date'] = pd.to_datetime(order_item_df['shipping_limit_date'])
order_item_df['bulan_tahun'] = order_item_df['shipping_limit_date'].dt.to_period('M')
order_item_df['total_penjualan_per_bulan'] = order_item_df.groupby('bulan_tahun')['freight_value'].transform('sum')
order_item_df = order_item_df.sort_values('bulan_tahun')
print(order_item_df[['bulan_tahun', 'total_penjualan_per_bulan']])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=order_item_df['bulan_tahun'].astype(str), y=order_item_df['total_penjualan_per_bulan'], marker='o')
plt.title('Grafik Tren Penjualan per Bulan')
plt.xlabel('Bulan_Tahun')
plt.ylabel('Total Penjualan per Bulan')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bulan_tahun', y='freight_value', data=order_item_df)
plt.title('Distribusi Nilai Penjualan per Bulan')
plt.xlabel('Bulan_Tahun')
plt.ylabel('Nilai Penjualan')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Pertanyaan 2:

```python
frekuensi_state = customers_df['customer_state'].value_counts().reset_index()

# Ubah nama kolom
frekuensi_state.columns = ['customer_state', 'frekuensi']

# Buat visualisasi menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='customer_state', y='frekuensi', data=frekuensi_state, palette='viridis')
plt.title('Frekuensi Daerah Asal Pelanggan')
plt.xlabel('Daerah (State)')
plt.ylabel('Frekuensi')
plt.show()
```
